<h4>Now , Deal with cleaning and analysis the dataset</h4>

```python
# STEP 1 Standardize column names (remove leading/trailing spaces)
for df in [order, sku_master, pincode, invoice, courier_rates]:
    df.columns = df.columns.str.strip()
```

```python
# STEP 2: Merge Order Report with SKU Master to get product weights
order_details = order.merge(sku_master, on="SKU", how="left")
order_details["Total Weight (kg)"] = (order_details["Weight (g)"] * order_details["Order Qty"]) / 1000
```

```python
# STEP 3: Calculate total weight per Order ID
order_weight = order_details.groupby("ExternOrderNo")["Total Weight (kg)"].sum().reset_index()
order_weight.rename(columns={
    "ExternOrderNo": "Order ID",
    "Total Weight (kg)": "Total Weight as per X (KG)"
}, inplace=True)
```

```python
order_weight.head()
```

```python
# STEP 4: Apply weight slab logic
def get_weight_slab(weight):
    return np.ceil(weight * 2) / 2

order_weight["Weight slab as per X (KG)"] = order_weight["Total Weight as per X (KG)"].apply(get_weight_slab)
```

```python
# STEP 5: Map delivery zones using pincode data
pincode_info = invoice[["AWB Code", "Warehouse Pincode", "Customer Pincode"]].copy()
pincode_info = pincode_info.merge(pincode, on=["Warehouse Pincode", "Customer Pincode"], how="left")
pincode_info.rename(columns={"Zone": "Delivery Zone as per X"}, inplace=True)
```

```python
pincode_info
```

```python
# STEP 6: Merge invoice with weights and zone info
invoice.rename(columns={"ExternOrderNo": "Order ID"}, inplace=True)  # Align order ID 
full_data = invoice.merge(order_weight, on="Order ID", how="left")
full_data = full_data.merge(pincode_info[["AWB Code", "Delivery Zone as per X"]], on="AWB Code", how="left")
```

```python
# STEP 7: Courier company charged weight slab
full_data["Weight slab charged by Courier Company (KG)"] = full_data["Charged Weight"].apply(get_weight_slab)
```

```python
# STEP 8: Expected charge calculation using courier rate card
rate_row = courier_rates.iloc[0]  # Assuming single row rate card

def calculate_expected_charge(row):
    slab = row["Weight slab as per X (KG)"]
    zone = row["Delivery Zone as per X"]
    shipment_type = row["Type of Shipment"]

    if pd.isnull(slab) or pd.isnull(zone):
        return np.nan

    try:
        fixed = rate_row[f"fwd_{zone}_fixed"]
        additional = rate_row[f"fwd_{zone}_additional"]
    except KeyError:
        return np.nan

    add_units = max(0, int((slab - 0.5) / 0.5))
    expected = fixed + (add_units * additional)

    if shipment_type.lower().strip() == "forward and rto charges":
        rto_fixed = rate_row[f"rto_{zone}_fixed"]
        rto_additional = rate_row[f"rto_{zone}_additional"]
        expected += rto_fixed + (add_units * rto_additional)

    return round(expected, 2)

full_data["Expected Charge as per X (Rs.)"] = full_data.apply(calculate_expected_charge, axis=1)
```

```python
full_data
```

```python
# STEP 9: Difference between expected and actual billed charges
full_data["Difference Between Expected Charges and Billed Charges (Rs.)"] = \
    full_data["Expected Charge as per X (Rs.)"] - full_data["Billing Amount (Rs.)"]
```

```python
# STEP 10: Final cleaned report
final_output = full_data[[
    "Order ID", "AWB Code",
    "Total Weight as per X (KG)", "Weight slab as per X (KG)",
    "Charged Weight", "Weight slab charged by Courier Company (KG)",
    "Delivery Zone as per X", "Zone",
    "Expected Charge as per X (Rs.)", "Billing Amount (Rs.)",
    "Difference Between Expected Charges and Billed Charges (Rs.)"
]].rename(columns={
    "Charged Weight": "Total weight as per Courier Company (KG)",
    "Zone": "Delivery Zone charged by Courier Company"
})
```

```python
final_output.head(20)
```

```python
# STEP 11: Summary analysis
summary = {
    "Total orders where X has been correctly charged": final_output[
        final_output["Difference Between Expected Charges and Billed Charges (Rs.)"] == 0
    ].shape[0],

    "Total Orders where X has been overcharged": final_output[
        final_output["Difference Between Expected Charges and Billed Charges (Rs.)"] < 0
    ].shape[0],

    "Total Orders where X has been undercharged": final_output[
        final_output["Difference Between Expected Charges and Billed Charges (Rs.)"] > 0
    ].shape[0],

    "Total overcharging amount": final_output[
        final_output["Difference Between Expected Charges and Billed Charges (Rs.)"] < 0
    ]["Difference Between Expected Charges and Billed Charges (Rs.)"].abs().sum(),

    "Total undercharging amount": final_output[
        final_output["Difference Between Expected Charges and Billed Charges (Rs.)"] > 0
    ]["Difference Between Expected Charges and Billed Charges (Rs.)"].sum(),

    "Total invoice amount": final_output["Billing Amount (Rs.)"].sum()
}

summary_df = pd.DataFrame.from_dict(summary, orient="index", columns=["Value"])
```

<h3>8)  Billing accuracy distribution </h3>

```python

# Add a Billing Accuracy column for analysis
def get_billing_accuracy(row):
    diff = row["Difference Between Expected Charges and Billed Charges (Rs.)"]
    if diff == 0:
        return "Correctly Charged"
    elif diff < 0:
        return "Overcharged"
    else:
        return "Undercharged"

final_output["Billing Accuracy"] = final_output.apply(get_billing_accuracy, axis=1)

# Also add slab match flag
final_output["Slab Match"] = final_output.apply(
    lambda row: "Match" if row["Weight slab as per X (KG)"] == row["Weight slab charged by Courier Company (KG)"]
    else "Mismatch", axis=1
)
accuracy_counts = final_output["Billing Accuracy"].value_counts()
#visualize billing accuracy distribution 
plt.figure(figsize=(6, 6))
colors = ["#4CAF50", "#F44336", "#FFC107"]
plt.pie(accuracy_counts, labels=accuracy_counts.index, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Billing Accuracy Distribution")
plt.tight_layout()
plt.show()

```
